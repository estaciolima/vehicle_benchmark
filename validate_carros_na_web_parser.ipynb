Notebook de teste para verificar o parser em dois PDFs do mesmo tipo: Corolla Cross e Taos.

A ideia aqui ? validar se a l?gica geral continua funcionando em arquivos parecidos, mesmo com diferen?as de nomes e quantidade de linhas.

```python
from pathlib import Path

from carros_na_web_parser import parse_pdf, parse_pdf_with_sections
```

```python
pdfs = [
    next(Path(".").glob("*Toyota Corolla Cross*.pdf")),
    next(Path(".").glob("*Volkswagen Taos*.pdf")),
]
pdfs
```

```python
results = {}
for pdf in pdfs:
    df = parse_pdf(pdf)
    results[pdf.name] = df
    print(f"=== {pdf.name} ===")
    print(f"rows: {len(df)}")
    print(df.head(15).to_string(index=False))
    print("--- tail ---")
    print(df.tail(8).to_string(index=False))
    print()
```

```python
# Checagem r?pida de alguns campos-chave
expected_labels = ["Ano", "Pre?o", "Propuls?o", "Proced?ncia", "Garantia"]
for name, df in results.items():
    print(f"=== {name} ===")
    mask = df["Atributo veicular"].isin(expected_labels)
    print(df.loc[mask].to_string(index=False))
    print()
```

```python
# Vers?o com se??o auxiliar para depura??o
for pdf in pdfs:
    df_sections = parse_pdf_with_sections(pdf)
    print(f"=== {pdf.name} (with sections) ===")
    print(df_sections.head(12).to_string(index=False))
    print(f"rows: {len(df_sections)}")
    print()
```
